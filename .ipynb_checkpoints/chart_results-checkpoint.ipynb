```python
LR_PALLETTE = "#FFB931"
CNN_PALLETTE = "#CCDDEF"
POLY_PALLETTE = "#1E2E4D"
BAG2012_PALLETTE = "#9CC39A"
```

```python
# set ticks on x-axis
CAND_SIZES = list(map(int, np.linspace(5, 40, 8)))
CAND_SIZES_OBFUSCATION = list(map(int, np.linspace(5, 25, 5)))  # 27
CAND_SIZES_IMITATION = list(map(int, np.linspace(5, 15, 3)))  # 17
CAND_SIZES_CONTROL = list(map(int, np.linspace(5, 20, 4)))  # 21
CAND_SIZES_SPECIAL_ENGLISH = list(map(int, np.linspace(5, 15, 3)))  # 18
CAND_SIZES_TRANSLATION = CAND_SIZES_CONTROL  # 21
```

```python
# data from orginal paper
ebg_cv_bag2012 = list(pd.read_csv('resource/webplotdigitizer/bag2012_fig6.csv', names=['_', 'acc']).acc)
ebg_obfuscation_bag2012 = list(pd.read_csv('resource/webplotdigitizer/bag2012_fig7.csv', names=['_', 'acc']).acc)
ebg_imitation_bag2012 = list(pd.read_csv('resource/webplotdigitizer/bag2012_fig8.csv', names=['_', 'acc']).acc)
# ebg, poly svm
ebg_imi_svm = json.load(open('results/ebg_imitation_svm.json', 'rb'))
ebg_obf_svm = json.load(open('results/ebg_obfuscation_svm.json', 'rb'))
ebg_cv_svm = json.load(open('results/ebg_cross_validation_svm.json', 'rb'))
```

```python
# rj, svm-poly
rj_imi_svm = json.load(open('results/rj_imitation_svm.json', 'rb'))
rj_obf_svm = json.load(open('results/rj_obfuscation_svm.json', 'rb'))
rj_control_svm = json.load(open('results/rj_control_svm.json', 'rb'))
rj_trans_de_svm = json.load(open('results/rj_backtranslation_de_svm.json', 'rb'))
rj_trans_ja_svm = json.load(open('results/rj_backtranslation_ja_svm.json', 'rb'))
rj_trans_de_ja_svm = json.load(open('results/rj_backtranslation_de_ja_svm.json', 'rb'))
rj_cv_svm = json.load(open('results/rj_cross_validation_svm.json', 'rb'))
# rj, logistic regression
rj_imi_lr = json.load(open('results/rj_imitation_logistic_regression.json', 'rb'))
rj_obf_lr = json.load(open('results/rj_obfuscation_logistic_regression.json', 'rb'))
rj_control_lr = json.load(open('results/rj_control_logistic_regression.json', 'rb'))
rj_trans_de_lr = json.load(open('results/rj_backtranslation_de_logistic_regression.json', 'rb'))
rj_trans_ja_lr = json.load(open('results/rj_backtranslation_ja_logistic_regression.json', 'rb'))
rj_trans_de_ja_lr = json.load(open('results/rj_backtranslation_de_ja_logistic_regression.json', 'rb'))
rj_cv_lr = json.load(open('results/rj_cross_validation_logistic_regression.json', 'rb'))
```

# Plot Reproduction

```python
# returns confidence interval of mean
def ci(a, conf=0.95):
    sem, m = st.sem(a), st.t.ppf((1+conf)/2., len(a)-1)
    return m*sem
```

```python
# reproduce BAG2012 Fig. 6

# step 1. get control of everything
width = 6.3
height = width / 2.618
barWidth = 1.1
plt.rc('font', family='serif', serif='Times')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)
plt.rc('axes', labelsize=8)
fig=plt.figure(figsize=(width, height), dpi=1200)
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# step 2. plot baselines (at ax1) and accuracy after including online documents (at ax2) for comparison

#fig6#########################################################################################################
# svm
ax1.bar(x=[x-.5*barWidth for x in CAND_SIZES],
        height=list(map(np.mean, ebg_cv_svm.values())),
        yerr= list(map(ci, ebg_cv_svm.values())),
        error_kw={'elinewidth':.3, 'capsize':.3},
        color=POLY_PALLETTE,
        linestyle='--',
        label='Writeprints-Static (Polynomial SVM)',
        width=barWidth)
# bag2012
ax1.bar([x+.5*barWidth for x in CAND_SIZES],
        ebg_cv_bag2012,
        color=BAG2012_PALLETTE,
        linestyle='--',
        label="Writeprints-Static (Polynomial SVM) (Brennan et al. 2012)",
        width=barWidth)
# chance
ax1.plot(CAND_SIZES,
         [1/c for c in CAND_SIZES],
         color='black',
         linestyle= '-',
         label='Random',
         linewidth=0.75)

# step 3. fine tune other parts with trail and error
ax1.set_title('Cross-validation on Training', {'fontsize':8})
ax1.set_xlabel('Number of Authors')
ax1.set_ylabel('Accuracy')
ax1.set_xlim(0, 45)
ax1.set_xticks(CAND_SIZES, minor=False)
ax1.set_ylim(0, 1)
ax1.grid(which='both', axis='y')

#########################################################################################################
#fig7########################################################################################################

# chance
ax2.plot(CAND_SIZES,
         [1/c for c in CAND_SIZES],
         color='black',
         linestyle= '-',
         label='Random',
         linewidth=0.75)
# svm
ax2.bar(x=[x-.5*barWidth for x in CAND_SIZES],
        height=list(map(np.mean, ebg_obf_svm.values())),
        yerr= list(map(ci, ebg_obf_svm.values())),
        error_kw={'elinewidth':.3, 'capsize':.3},
        color=POLY_PALLETTE,
        linestyle='--',
        label='Writeprints-Static (Polynomial SVM)',
        width=barWidth)
# bag2012
ax2.bar([x+.5*barWidth for x in CAND_SIZES],
        ebg_obfuscation_bag2012,
        color=BAG2012_PALLETTE,
        linestyle='--',
        label="Writeprints-Static (Polynomial SVM) (Brennan et al. 2012)",
        width=barWidth)


# step 3. fine tune other parts with trail and error
ax2.set_title('Obfuscation', {'fontsize':8})
ax2.set_xlabel('Number of Authors')
# ax2.set_ylabel('Accuracy')
ax2.set_xlim(0, 45)
ax2.set_xticks(CAND_SIZES, minor=False)
ax2.set_ylim(0, 1)
ax2.grid(which='both', axis='y')
#########################################################################################################
#fig8########################################################################################################

# chance
ax3.plot(CAND_SIZES,
         [1/c for c in CAND_SIZES],
         color='black',
         linestyle= '-',
         label='Random',
         linewidth=0.75)
# poly
ax3.bar(x=[x-.5*barWidth for x in CAND_SIZES],
        height=list(map(np.mean, ebg_imi_svm.values())),
        yerr= list(map(ci, ebg_imi_svm.values())),
        error_kw={'elinewidth':.3, 'capsize':.3},
        color=POLY_PALLETTE,
        linestyle='--',
        label='Writeprints-Static (Polynomial SVM)',
        width=barWidth)
# bag2012
ax3.bar([x+.5*barWidth for x in CAND_SIZES],
        ebg_imitation_bag2012,
        color=BAG2012_PALLETTE,
        linestyle='--',
        label="Writeprints-Static (Polynomial SVM) (Brennan et al. 2012)",
        width=barWidth)


# step 3. fine tune other parts with trail and error
ax3.set_title('Imitation', {'fontsize':8})
ax3.set_xlabel('Number of Authors')
# ax3.set_ylabel('Accuracy')
ax3.set_xlim(0, 45)
ax3.set_xticks(CAND_SIZES, minor=False)
ax3.set_ylim(0, 1)
ax3.grid(which='both', axis='y')

#########################################################################################################
#########################################################################################################
handles, labels = ax1.get_legend_handles_labels()
# # labels = ["\n".join(wrap(label, 15)) for label in labels]
# lgd = ax2.legend(handles,
#                  ["\n".join(wrap(label, 20)) for label in labels],
#                  loc='right',
#                  bbox_to_anchor=(1, -0.5, 0.5, 0.5),
#                  fontsize='xx-small')
# labels = ["\n".join(wrap(label, 15)) for label in labels]
plt.figlegend(handles,
              labels,
#               bbox_to_anchor=(0.92, 0.2, 0.2, 0.5), # side-best
              bbox_to_anchor=(0.7, -0.48, 0.2, 0.5), # bottom
              ncol=3,
              labelspacing=0,
             fontsize='xx-small')
# fig.legend()

# add a convenient function to transfer y-ticks from decimal to percentage, for better readability
def to_percent(temp, position):
    return '%1.0f'%(100*temp) + '\%'

# reformat 
ax1.yaxis.set_major_formatter(FuncFormatter(to_percent))
ax2.set_yticklabels([])
ax3.set_yticklabels([])
# ax2.yaxis.set_major_formatter(FuncFormatter(to_percent))
# ax3.yaxis.set_major_formatter(FuncFormatter(to_percent))

# remove unsolicited spines to increase data-ink ratio
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)

# remove tick markers
ax1.tick_params(left = False, bottom = False)
ax2.tick_params(left = False, bottom = False)
ax3.tick_params(left = False, bottom = False)

# adjust grid lines
ax1.grid(visible=True, linestyle='-', axis='y', which='major', linewidth=0.5, alpha=0.5)
ax2.grid(visible=True, linestyle='-', axis='y', which='major', linewidth=0.5, alpha=0.5)
ax3.grid(visible=True, linestyle='-', axis='y', which='major', linewidth=0.5, alpha=0.5)

# plt.suptitle('Reproduction of Adversarial Stylometry on the Extended Brennan-Greenstadt Corpus', fontsize=9)

plt.tight_layout()

plt.savefig('figures/reproduction_summary.png', bbox_inches='tight')

plt.show()
```
